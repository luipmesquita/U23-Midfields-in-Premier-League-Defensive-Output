```python
import pandas as pd
import soccerdata as sd

fbref = sd.FBref(leagues="ENG-Premier League", seasons="25-26")
```

```python
df_playing_time = fbref.read_player_season_stats(stat_type="playing_time")
df_misc = fbref.read_player_season_stats(stat_type="misc")
```

```python
print(df_playing_time)
```

```python
# Transforming multindex in normal index
df_playing_time = df_playing_time.reset_index()
print(df_playing_time.columns.tolist())
df_playing_time.head()
```

```python
# Transforming multindex in normal index
df_misc = df_misc.reset_index()
print(df_misc.columns.tolist())
df_misc.head()
```

```python
df_scouting = pd.merge(df_playing_time, df_misc, on=['league', 'season', 'team', 'player'], how='inner')
```

```python
print(df_scouting.columns.tolist())
print(df_scouting.head())
```

```python
# Removing tuples from the array
df_scouting.columns = [
    f"{col[0]}_{col[1]}" if isinstance(col, tuple) and col[1] != "" 
    else col[0] if isinstance(col, tuple) 
    else col
    for col in df_scouting.columns
]
df_scouting = df_scouting.loc[:, ~df_scouting.columns.str.endswith('_y')]
df_scouting.columns = df_scouting.columns.str.replace('_x', '')
```

```python
print(df_scouting.columns.tolist())
df_scouting
```

```python
# Filter position, age, playing time e playing time per matches
df_anchor_man = df_scouting[df_scouting['pos'].str.contains('MF', na=False, case=False)]
```

```python
df_anchor_man = df_anchor_man[~df_anchor_man['pos'].str.contains('FW', na=False, case=False)]
```

```python
df_anchor_man = df_anchor_man[df_anchor_man['age'] <= 23]
```

```python
df_anchor_man = df_anchor_man[df_anchor_man['Playing Time_Min'] >= 500]
```

```python
df_anchor_man = df_anchor_man[df_anchor_man['Playing Time_Mn/MP'] >= 45]
```

```python
print(f"Found {df_anchor_man.shape[0]} players")
```

```python
df_anchor_man
```

```python
df_anchor_man['Def_Actions_90'] = (df_anchor_man['Performance_TklW'] + df_anchor_man['Performance_Int']) / (df_anchor_man['Playing Time_Min'] / 90)
```

```python
df_anchor_man['Team_GA_90'] = df_anchor_man['Team Success_onGA'] / df_anchor_man['Playing Time_90s']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Graph style
sns.set_theme(style="white")
plt.figure(figsize=(13, 8))

# Scatter plot with axis
ax = sns.scatterplot(
    data=df_anchor_man, 
    x='Def_Actions_90', 
    y='Team_GA_90', 
    s=120, 
    color='#0284c7',
    edgecolor='black',
    linewidth=0.8,
    alpha=0.9
)

# Custom gridlines
plt.grid(visible=True, which='major', axis='both', color='#e5e7eb', linestyle=':', linewidth=0.8)

# Mean lines to have 4 quadrants
plt.axvline(df_anchor_man['Def_Actions_90'].mean(), color='#ef4444', linestyle='--', alpha=0.5, linewidth=1.2, label='Defensive Actions Mean')
plt.axhline(df_anchor_man['Team_GA_90'].mean(), color='#10b981', linestyle='--', alpha=0.5, linewidth=1.2, label='Goals Conceded Mean')

# Add players next to dots
for i in range(df_anchor_man.shape[0]):
    plt.text(
        df_anchor_man['Def_Actions_90'].iloc[i] + 0.04, 
        df_anchor_man['Team_GA_90'].iloc[i], 
        df_anchor_man['player'].iloc[i], 
        fontsize=8,
        fontweight='medium',
        color='#1f2937',
        alpha=0.85,
        va='center'
    )
    
# Filters notes, top corner
plt.text(
    1.0, 1.04, 
    'Players with >45 min/game and >=500 total min', 
    fontsize=9, 
    color='#6b7280', 
    fontstyle='italic',
    ha='right',
    transform=ax.transAxes
)

# Invert y-axis
plt.gca().invert_yaxis()

# Title, labels and legends
plt.suptitle('U23 Midfielders in Premier League', fontsize=16, fontweight='bold', color='#111827', y=0.96)
plt.title('Team vs Individual Defensive Output', fontsize=11, fontweight='medium', fontstyle='italic', color='#111827', pad=15)
plt.xlabel('Defensive Actions (Tackles Won + Interceptions) per 90 minutes', fontsize=11, fontweight='medium', color='#374151')
plt.ylabel('Goals Conceded per 90 minutes (Inverted)', fontsize=11, fontweight='medium', color='#374151')
plt.legend(loc='upper left', frameon=True, facecolor='white', edgecolor='#e5e7eb')

sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
